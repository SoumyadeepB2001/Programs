```python
#Training Dataset
train_data=pd.read_csv('gdrive/My Drive/Files/Acoustic_train.csv', index_col=False, engine='python')
print(train_data.shape)
train_data.head()
```

```python
train_data.info()
```

```python
#Test Dataset
test_data=pd.read_csv('gdrive/My Drive/Files/Acoustic_test_no_class.csv', index_col=False, engine='python')
print(test_data.shape)
test_data.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Normalize features
scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(train_data.drop(columns=['Index','Class']))
test_data_scaled = scaler.transform(test_data.drop(columns=['Index']))
```

```python
train_data_scaled
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode labels
label_encoder = LabelEncoder()
train_labels = label_encoder.fit_transform(train_data['Class'])
```

```python
train_labels
```

```python

from tensorflow import keras

# Build Deep Learning model
model = keras.Sequential([
    keras.layers.Input(shape=(train_data_scaled.shape[1],)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(len(label_encoder.classes_), activation='softmax')
])

# Compile model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Display model summary
model.summary()
```

```python
# Train the model
model.fit(train_data_scaled, train_labels, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# Predict on test data
y_test_pred = model.predict(test_data_scaled)
y_test_labels = y_test_pred.argmax(axis=1)
test_data["Predicted_Class"] = label_encoder.inverse_transform(y_test_labels)

# Save predictions to CSV
test_data.to_csv("Acoustic_test_predictions.csv", index=False)

```

```python
from sklearn.metrics import classification_report, accuracy_score

# Predict on trining data
y_train_pred = model.predict(train_data_scaled).argmax(axis=1)

# Print classification report
print(classification_report(train_labels, y_train_pred, target_names=label_encoder.classes_))
```

```python
#Training accuracy
train_loss, train_accuracy = model.evaluate(train_data_scaled, train_labels )
print(f"Training Accuracy: {train_accuracy * 100:.2f}%")
```

```python
# Predict on test data
y_test_pred = model.predict(test_data_scaled)
y_test_labels = y_test_pred.argmax(axis=1)
test_data["Predicted_Class"] = label_encoder.inverse_transform(y_test_labels)

# Save predictions to CSV
test_data.to_csv("Acoustic_test_predictions.csv", index=False)
```

```python
y_test_true = test_data["Predicted_Class"]
y_test_true_encoded = label_encoder.transform(y_test_true)

# Convert predictions to class labels
y_pred_labels = model.predict(test_data_scaled).argmax(axis=1)

# Compute accuracy by comparing the true labels and the predicted labels
test_accuracy = accuracy_score(y_test_true_encoded, y_pred_labels)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
